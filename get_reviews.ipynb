# Importing libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from time import sleep
from random import randint
```

# Defining functions

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'}
```

Function to get soup from url

```python
def get_data(url):
    request = requests.get(url, headers=headers)
    soup = BeautifulSoup(request.text,"html.parser")
    return soup
```

Function to assemble link to product page

```python
def asin_url(asins):
    url = 'https://www.amazon.co.uk/dp/' + asins
    print(url)
    page = requests.get(url,headers=headers)
    if page.status_code==200:
        return page
```

Function to get review pages

```python
def search_reviews(review_link):
    url = 'https://www.amazon.co.uk' + review_link
    print(url)
    page=requests.get(url,headers=headers)
    if page.status_code==200:
        return page
```

Function to scrape reviews

```python
def get_reviews(soup):
    reviews = soup.find_all('div', {'data-hook': 'review'})
    try:
        for item in reviews:
            review = {
            'product': soup.title.text.strip(),
            'date': item.find('span', {'data-hook': 'review-date'}).text.strip(),
            'title': item.find('a', {'data-hook': 'review-title'}).text.strip(),
            'rating': item.find('i', {'data-hook': 'review-star-rating'}).text.strip(),
            'body': item.find('span', {'data-hook': 'review-body'}).text.strip(),
            }
            reviewlist.append(review)
    except:
        review = {
            'product': soup.title.text.strip(),
            'date': item.find('span', {'data-hook': 'review-date'}).text.strip(),
            'title': item.find('span', {'data-hook': 'review-title'}).text.strip(),
            'rating': item.find('i', {'data-hook': 'cmps-review-star-rating'}).text.strip(),
            'body': item.find('span', {'data-hook': 'review-body'}).text.strip(),
            }
        reviewlist.append(review)
```

# Getting data

Getting all overview pages from the item search

```python
pages = []

for i in range(1,5):
    start = str(i)
    url = 'https://www.amazon.co.uk/s?k=wooden+dolls+house&crid=11XBG7KGKSLO0&qid=1667196153&sprefix=wooden+d%2Caps%2C83&ref=sr_pg_' + start
    response = requests.get(url, headers = headers)
    pages.append(url)
```

```python
pages
```

Getting product names

```python
product_names = []

for link in pages: 
    response = requests.get(link, headers = headers)
    soup = BeautifulSoup(response.content,"html.parser")
    for i in soup.find_all('span',{'class': 'a-size-base-plus a-color-base a-text-normal'}):
            product_names.append(i.text.strip())
```

```python
len(product_names)
```

Getting product ID's 

```python
asins = []
for link in pages:
    response = requests.get(link, headers = headers)
    soup = BeautifulSoup(response.content,"html.parser")
    for x in soup.select('div[data-asin]'):
        if x['data-asin'] != '':
            asins.append(x['data-asin'])
```

```python
len(asins)
```

Getting links for all products

```python
link = []
for x in range(len(asins)):
    response = asin_url(asins[x])
    soup = BeautifulSoup(response.content, "html.parser")
    for x in soup.find_all("a",{'data-hook':"see-all-reviews-link-foot"}):
            link.append(x['href'])
```
